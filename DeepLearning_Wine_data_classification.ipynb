```python
import tensorflow as tf
import pandas as pd      ##for csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
red = pd.read_csv("winequality-red.csv" , sep=";")
white = pd.read_csv("winequality-white.csv" , sep=";")
```

```python
# concat 2 datasets and merge them
## first def class

red["Type"] = 1
white["Type"] = 0
```

```python
##append the data in white from red and to ignore index for concatinating the dataset
wines = white.append(red, ignore_index=True)
```

```python
wines.shape
```

```python
wines
```

```python
## split and shuffle data and apply StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

##cols independent and type dependent so diff datasets req
x,y = wines.iloc[:, :-1].values, wines.iloc[:, -1].values
trainX, testX, trainY, testY = train_test_split(x,y, test_size = 0.2, random_state = 12)
```

```python
## convert data i.e scalardataset
scaller = StandardScaler()
trainX = scaller.fit_transform(trainX)
testX = scaller.fit_transform(testX)
```

# Deep Learning

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
trainX.shape
```

```python
model = Sequential()    ##means object created from sequential class and we can store anything
```

```python
model
```

```python
## add layers based on input
model.add(Dense(30, activation = "relu", input_shape = (12, )))   ##input shape is trainX shape or independent columns
model.add(Dense(50, activation = "relu"))     ##to avoid underfit we add layers
model.add(Dense(1, activation = "sigmoid"))  ##sigmoid for probablity
model.compile(loss = "binary_crossentropy", optimizer="adam", metrics=["accuracy"])  ##building model done now compile ie loss and to compile loss we need binary_crossentropy
history = model.fit(trainX, trainY, validation_split = 0.3, epochs = 20, batch_size = trainX.shape[0])
##fir our model and store ie history and then validate data by validation split and then epochs is no.of times u execute/ iterate ur algo and batch size has neurons u want to pass ie trainX shape
```

```python
model.summary()   ##check for layers   
## we have 12 inputs and 30 inputs ie 360 and every neuron has bias so 360+30
##on exectuing the n times it gives two times the layers present 
```

```python
history.history
```

```python
loss = history.history["loss"]
val_loss = history.history["val_loss"]
accuracy = history.history["accuracy"]
val_accuracy = history.history["val_accuracy"]
```

```python
tva = pd.DataFrame({              ##training validation accuracy
    "TrainingAcc":accuracy,
    "Val_Acc":val_accuracy
})
```

```python
tva.boxplot()
```
